```python
## Constructing Binary Tree - Task 1

# this basically symbolizes the node that a binary tree has
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root_data):
        self.root = Node(root_data)  # this intilizes the root node
    
    # function to insert left data in the binary tree
    def insert_left(self, parent_node, data):
        if parent_node.left is None:
            parent_node.left = Node(data)
        else:
            new_node = Node(data)
            new_node.left = parent_node.left
            parent_node.left=new_node

    #function to insert right in the binary tree
    def insert_right(self, parent_node, data):
        if parent_node.right is None:
            parent_node.right = Node(data)
        else:
            new_node = Node(data)
            new_node.right = parent_node.right
            parent_node.right = new_node

    # building the BFS

    def bfs(self, root):
        if not root:
            return

        result = []
        queue = [root]

        while queue:
            node = queue.pop(0)
            result.append(node.data)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        return result


    # building dfs using stack

    def dfs(self, root):
        if not root:
            return []

        result = []
        stack = [root] 

        while stack:
            node = stack.pop()
            result.append(node.data)

            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
        return result

```

```python
from binarytree import tree
import time
import matplotlib.pyplot as plt

# Your Node class (unchanged)
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root_data):
        self.root = Node(root_data)  # Initializes the root node
    
    # Insert left (fixed 'none' to 'None')
    def insert_left(self, parent_node, data):
        if parent_node.left is None:
            parent_node.left = Node(data)
        else:
            new_node = Node(data)
            new_node.left = parent_node.left
            parent_node.left = new_node

    # Insert right (fixed 'none' to 'None')
    def insert_right(self, parent_node, data):
        if parent_node.right is None:
            parent_node.right = Node(data)
        else:
            new_node = Node(data)
            new_node.right = parent_node.right
            parent_node.right = new_node

    # BFS with timing
    def bfs(self, root):
        start_time = time.perf_counter()  # Start timing
        if not root:
            end_time = time.perf_counter()  # End timing
            return [], end_time - start_time

        result = []
        queue = [root]

        while queue:
            node = queue.pop(0)
            result.append(node.value if hasattr(node, 'value') else node.data)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        end_time = time.perf_counter()  # End timing
        return result, end_time - start_time

    # DFS with timing
    def dfs(self, root):
        start_time = time.perf_counter()  # Start timing
        if not root:
            end_time = time.perf_counter()  # End timing
            return [], end_time - start_time

        result = []
        stack = [root]

        while stack:
            node = stack.pop()
            result.append(node.value if hasattr(node, 'value') else node.data)
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
        
        end_time = time.perf_counter()  # End timing
        return result, end_time - start_time

def plot_execution_times(heights=[3, 5, 7, 9]):
    # Lists to store data for plotting
    node_counts = []
    bfs_times = []
    dfs_times = []
    
    bt = BinaryTree(0)  # Dummy root
    
    for height in heights:
        # Generate random tree
        lib_root = tree(height=height)
        node_count = len(lib_root.values)  # Count nodes
        
        # Run BFS and DFS
        _, bfs_time = bt.bfs(lib_root)
        _, dfs_time = bt.dfs(lib_root)
        
        # Store results
        node_counts.append(node_count)
        bfs_times.append(bfs_time)
        dfs_times.append(dfs_time)
        
        # Print results for reference
        print(f"Height: {height}, Nodes: {node_count}, BFS Time: {bfs_time:.6f}s, DFS Time: {dfs_time:.6f}s")
    
    # # Plot using Matplotlib
    plt.plot(node_counts, bfs_times, marker='o', label='BFS', color='blue')
    plt.plot(node_counts, dfs_times, marker='s', label='DFS', color='red')
    plt.xlabel('Number of Nodes')
    plt.ylabel('Execution Time (seconds)')
    plt.title('BFS and DFS Execution Time vs. Tree Size')
    plt.grid(True)
    plt.legend()
    plt.show()

# Run the test and plot
if __name__ == "__main__":
    print("=== Execution Time Analysis ===")
    plot_execution_times(heights=[3, 5, 7, 9])
```
