```python
import pandas as pd

data = pd.read_csv("suiciderate.csv")
```

```python
# Visualizing Data on Terminal

print(data.head())
print(data.tail())
print(data.shape)
print(data.describe())
print(data.info())
print(data.isnull().sum())
```

```python
data.columns = data.columns.str.strip()
type_verification = data['gdp_per_capita ($)'].dtype
print(type_verification) # returns float64
data['gdp_per_capita ($)']=pd.to_numeric(data['gdp_per_capita ($)'])
type_verification_2 = data['gdp_for_year ($)'].dtype
if type_verification_2 == 'object':
  data['gdp_for_year ($)'] = data[' gdp_for_year ($) '].str.replace(',', '')
  data['gdp_for_year ($)'] = pd.to_numeric(data['gdp_for_year ($)'])
else:
  print("already a numeric") #returns this
data['suicide-to-GDP-ratio']=data['suicides_no']/data['gdp_for_year ($)']
data['suicide-to-GDP-ratio']
```

```python
age_grouped = data.groupby('age').agg({
    'suicides_no': 'sum',
    'population': 'sum'
})
age_grouped['suicide-to-population-ratio'] = age_grouped['suicides_no'] / age_grouped['population'] *100000
age_grouped['suicide-to-population-ratio']
```

```python
countries=data.groupby('country').agg({
    'suicides_no': 'sum',
    'population': 'sum'
})

countries['suicide-to-population-ratio'] = countries['suicides_no'] / countries['population'] *100
countries['suicide-to-population-ratio'].dropna()
countries['suicide-to-population-ratio'].sort_values(ascending=True)
countries
```

```python
gdp_grouped = data.groupby('gdp_for_year ($)').agg({
    'suicides_no': 'sum',
    'population': 'sum'
})
gdp_grouped['suicide_rate'] = (gdp_grouped['suicides_no'] / gdp_grouped['population']) * 100000
gdp_grouped = gdp_grouped.dropna()
gdp_grouped['suicide_rate'].sort_values(ascending=True)
```

```python
sex_grouped = data.groupby('sex').agg({
    'suicides_no': 'sum',
    'population': 'sum',
    'suicides/100k pop': 'mean'
}).reset_index()

sex_grouped['suicide_rate_calculated'] = (sex_grouped['suicides_no'] / sex_grouped['population']) * 100000

male_rate = sex_grouped.loc[sex_grouped['sex'] == 'male', 'suicide_rate_calculated'].values[0]
female_rate = sex_grouped.loc[sex_grouped['sex'] == 'female', 'suicide_rate_calculated'].values[0]
ratio = male_rate / female_rate
difference = male_rate - female_rate

difference
```

```python
generation_grouped = data.groupby('generation').agg({
    'suicides_no': 'sum',
    'population': 'sum',
})

generation_grouped['suicide_rate_calculated'] = (generation_grouped['suicides_no'] / generation_grouped['population']) * 100000
generation_grouped = generation_grouped.sort_values('suicide_rate_calculated', ascending=True)
max_row = generation_grouped['suicide_rate_calculated'].idxmax()
max_row
```
